## Grouping telescopes

```python
#group = {"LST": [1,2,3,4],
#        "MST": np.arange(5, 20)}

sub_groups={'1': np.array([5,17,18]), 
            '2': np.array([10,11,14]),
            '3': np.array([7,15,8]), 
            '4': np.array([9,13,19]),
            '5':np.array([6,12,16]),
            'LST': [1,2,3,4]}

array.display("xy", group=sub_groups)
```

### Subarray barycenter

```python
from ctadiv import COLORS
ax = array.display("xy", group=sub_groups)
groupped_table, labels = array.group_by(sub_groups)
for i, (tab, label) in enumerate(zip(groupped_table.groups, labels)):
    output = utils.calc_mean(tab, ["x", "y", "p_x", "p_y"])
    s = ax.scatter(output[0], output[1], color=COLORS(i),)
    ax.quiver(*output, color=s.get_facecolor())

plt.show(block=True)
```
