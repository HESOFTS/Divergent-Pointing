```python
from tqdm import tqdm
```

```python
mul_table = []
for div in tqdm(divs):
    for alt in [10, 30, 50, 70]:
        for az in [30, 60, 90, 120, 150, 180]:
            array.divergent_pointing(div, alt_mean = alt*u.deg, az_mean = az*u.deg)
            mul = utils.calc_multiplicity(array)
            mul_ave = np.average(mul[:,0], weights=mul[:,1])
            mul_table.append([div, alt, az, mul_ave])
            
mul_table = np.asarray(mul_table)
```

```python
for alt in [10, 30, 50, 70, 90]:
    plt.plot(mul_table[:,0][mul_table[:,1] == alt], mul_table[:,3][mul_table[:,1] == alt])
plt.xscale("log")
```

```python
for az in [30, 60, 90 ,120, 150, 180]:
    plt.plot(mul_table[:,0][mul_table[:,2] == az], mul_table[:,3][mul_table[:,2] == az])
plt.xscale("log")
```
